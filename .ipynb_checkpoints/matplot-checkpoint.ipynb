```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data={
    "Salary":[12000,23400,39000,20000,31000,67000,55000,63800,45000,43000]
}
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
#line plot
plt.plot(df['Salary'],marker='o',color='green',linestyle=':')
plt.grid()
plt.title("Salary Trend")
plt.show()
```

```python
# Histogram 
plt.hist(df['Salary'],bins=7,color='orange')
plt.show()
```

```python
df['dept']=['IT','UI/UX','DEV','DEV','IT','DEV','Finance','Finance','UI/UX','UI/UX']
```

```python
df
```

```python
# pie chart
count=df['dept'].value_counts()
plt.pie(count,labels=count.index,autopct='%1.1f')
        # ,explode=[0,0,0.1,0]
plt.show()
```

```python
# countplot
# plt.bar(count.index,count,color='green')
plt.bar(count.index,count,color=['red','pink','orange','yellow'])
plt.show()
```

```python
df['age']=[23,33,33,31,45,46,46,67,55,51]
df
```

```python
# scatter plot
plt.scatter(df['age'],df['Salary'])
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# line chart
sort_age=df.sort_values('age')
plt.plot(sort_age['age'],df['Salary'])
plt.show()
```

```python
plt.bar(sort_age['age'],df['Salary'])
plt.show()
```

```python
# box plot - dept(create categories) vs salary 
IT_sal=df[df['dept']=="IT"]['Salary']
DEV_sal=df[df['dept']=="DEV"]['Salary']
Finance_sal=df[df['dept']=="Finance"]['Salary']
UIUX_sal=df[df['dept']=="UI/UX"]['Salary']

plt.boxplot([IT_sal,DEV_sal,Finance_sal,UIUX_sal],labels=['IT','DEV','Finance','UI/UX'])
plt.title('dept vs Salary')
plt.xlabel('dept')
plt.ylabel('Salary')
plt.savefig('boxplt1.png')
plt.show()
```

```python
#piechart -  shows amt being earn by each department
salary_by_dept=df.groupby('dept')['Salary'].sum()

plt.pie(salary_by_dept,labels=salary_by_dept.index,autopct='%1.1f')
plt.show()
```
