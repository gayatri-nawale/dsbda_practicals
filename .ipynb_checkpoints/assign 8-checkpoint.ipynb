```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
print("Shape  : ",df.shape)
print("\nDimensions:  ",df.ndim)
print("\nData type:\n",df.dtypes)
print("\nSize:   ",df.size)
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['embarked']=df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town']=df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.drop('deck', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Survival Count
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
# Survival by Gender
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.barplot(x='pclass',y='survived',data=df,errorbar=None,palette='YlGnBu', hue='pclass')
plt.title("Survival Rate by Passenger Class")
plt.xlabel('Passenger class')
plt.ylabel('Survived')
plt.show()
```

```python
# Histogram for Fare Distribution

plt.figure(figsize=(6,5))
sns.histplot(df['fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")

plt.show()
```

```python
print("Skewness of Fare:", df['fare'].skew())
# Skewness value is greater than 0, confirming right skewness.
```

```python
# Age Distribution
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show() 
```

```python
print("Skewness of Age:", df['age'].skew())
# slightly normally distributed.
```

```python
# Fare vs Survival
sns.barplot(x='survived', y='fare', data=df, errorbar=None,palette='coolwarm',hue='survived')
plt.title("Average Fare vs Survival")
plt.show()
```

```python
# Age vs Survival

sns.boxplot(x='survived', y='age', data=df)
plt.title("Age vs Survival")
plt.show()

# 👉 Insight: Younger passengers had slightly higher survival.
```

```python
Most passengers were young adults (20–40 years)
Very few passengers were very old (above 60)
Age distribution is slightly right-skewed
```

```python
# Fare distribution by class

plt.figure(figsize=(6, 4))
sns.boxplot(x='pclass', y='fare', hue='pclass', data=df, palette='Set2', legend=False)
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Fare Distribution by Passenger Class')
plt.yscale('log')  # Log scale for better visualization
plt.show()
```
