```python
import pandas as pd
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
# original document
data="Natural language processing is branch of artifical engineering that is used to interprete, understand and manipulate human language"
print("Original Text:\n",data)
```

```python
# token generation
tokens=word_tokenize(data)
print("Tokenization Output:\n",tokens)
```

```python
# POS tagging
pos=pos_tag(tokens)
print("POS Output: \n",pos)
```

```python
# Take all words from tokens, remove common useless words 
# and remove symbols/numbers, keep only meaningful words.

# Removal of stop words 
stop_words=set(stopwords.words('english'))
token_no_stop_words=[word for word in tokens if word.lower() not in stop_words and word.isalpha()]
print("Stopword removal :\n",token_no_stop_words)
```

```python
# Stemming
stemmer=PorterStemmer()
stem_words=[stemmer.stem(word) for word in token_no_stop_words]
print("Stemming Output\n",stem_words)
```

```python
# lemmitization
lemmatizer=WordNetLemmatizer()
lemmatized_words=[lemmatizer.lemmatize(word) for word in token_no_stop_words]
print("Lemmitization Output\n",lemmatized_words)
```

```python
# TFIDF calculation
docs=[
    data,
    'Machine learning is important part of artifical intelligence that enables computers to learn from data.'
]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)
```

```python
tfidf_cal=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
print("TF-IDF Calculation")
print(tfidf_cal)
```

```python
# Higher value → word is more important in that document
# Lower value → word is less important
# 0.00000 → word not present in that document

# Rows (0, 1) → Documents
```
