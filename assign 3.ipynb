```python
import pandas as pd
```

```python
df=pd.read_csv('Iris.csv')
print(df.head())
print("Unique columns : ",df['Species'].unique())
```

```python
df.drop('Id',axis=1,inplace=True)
```

```python
print(df.head())
```

```python
print('Shape: ',df.shape)
print('\nDimension : ',df.ndim)
print('\nTotal number of elements : ',df.size)
print('\nData types: ',df.dtypes)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
# Compute mean, median, min, max, variance and standard deviation 
print("Mean :    \n",df.mean(numeric_only=True))
print("\nMedian :    \n",df.median(numeric_only=True))
print("\nMax value :    \n",df.max(numeric_only=True))
print("\nMin value :    \n",df.min(numeric_only=True))
print("\nVariance :    \n",df.var(numeric_only=True))
print("\nStandard Deviation :    \n",df.std(numeric_only=True))
```

```python
# Group data by categorical variable 

grp_species=df.groupby('Species')
print(grp_species.mean(numeric_only=1))
```

```python
print(grp_species.median(numeric_only=1))
```

```python
# Group data by categorical variable 
df.groupby('Species').agg(['mean','min','max'])
```
