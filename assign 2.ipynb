```python
df['DSBDA_score']=df['DSBDA_score'].fillna(df['DSBDA_score'].mean())
df['OOP_score']=df['OOP_score'].fillna(df['OOP_score'].mean())
```

```python
#detect the outlier
q1=df['DSBDA_score'].quantile(0.25) # gives value below which 25% data lies
q3=df['DSBDA_score'].quantile(0.75) # gives value below which 75% data lies

IQR=q3-q1 # spread of middle data

lower_limit=q1-1.5*IQR
upper_limit=q3+1.5*IQR
print("DSBDA range:", lower_limit, upper_limit)
```

```python
# handle the outlier Replace Outliers 
# (Outlier condition: value < 38.5  OR  value > 112.5)

# If value > upper → replaced with upper
# If value < lower → replaced with lower
# Normal values → remain same

df['DSBDA_score']=np.where(df['DSBDA_score']>upper_limit,upper_limit,df['DSBDA_score'])
df['DSBDA_score']=np.where(df['DSBDA_score']<lower_limit,lower_limit,df['DSBDA_score'])
```

```python
df # (manav- dsbda replace with upper limit)
```

```python
# OOP
q1= df['OOP_score'].quantile(0.25)
q3 = df['OOP_score'].quantile(0.75)

IQR = q3-q1

lower_limit = q1 - 1.5 * IQR
upper_limit = q3+ 1.5 * IQR

print("OOP range:", lower_limit, upper_limit)
```

```python
df['OOP_score'] = np.where(df['OOP_score'] > upper_limit, upper_limit, df['OOP_score'])
df['OOP_score'] = np.where(df['OOP_score'] < lower_limit, lower_limit, df['OOP_score'])
```

```python
df # (priya- oop replace with lower limit)
```

Apply Transformation 

```python
# log Tranformation
df['DSBDA_log']=np.log(df['DSBDA_score'])
print(df)
```

```python
# normalization
df['DSBDA_norm']=(df['DSBDA_score'] - df['DSBDA_score'].min()) / (df['DSBDA_score'].max() - df['DSBDA_score'].min())
print(df)
```

```python
#  Standarization
mean_value = df['DSBDA_score'].mean()
print(mean_value)

# Values around mean → near 0 
# Above mean → positive 
# Below mean → negative

df['DSBDA_scaled'] = (df['DSBDA_score'] - df['DSBDA_score'].mean()) / df['DSBDA_score'].std()
print(df)
```
